## Data Wrangling / Feature Engineering

### Identifying missing values

```python
df.head(10)
```

```python
# replace "?" to NaN
df.replace("?", np.nan, inplace = True)
```

```python
df.head()
```

### Evaluating for missing values

```python
missing_data = df.isnull()
missing_data.head(5)
```

```python
for column in missing_data.columns:
    print (missing_data[column].value_counts())
    print("")  
```

```python
# replacing the missing data with mean for numeric columns
avg_norm_loss = df["normalized-losses"].astype(float).mean(axis = 0)
avg_norm_loss
```

```python
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace = True)
```

```python
df.count()
```

```python
avg_bore=df['bore'].astype('float').mean(axis=0)
avg_bore
```

```python
df.head()
```

```python
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)

df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
avg_peakrpm
```

```python
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
# replacing the missing values in the categorical column of num-of-doors with the most frequent value

df['num-of-doors'].value_counts()
```

```python
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
df.head()
```

### Changing Data Format

```python
df.dtypes
```

```python
df[["bore", "stroke", "price", "peak-rpm"]] = df[["bore", "stroke", "price", "peak-rpm"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
```

```python
df.dtypes
```

```python
df["city-L/100km"] = 235/df["city-mpg"]

df.head()
```

```python
df["highway-L/100km"] = 235/df["highway-mpg"]

df.head()
```

## Multiple Linear Regression Model

```python
drive_wheel_num = pd.get_dummies(df["drive-wheels"])
drive_wheel_num = drive_wheel_num.astype(int)
drive_wheel_num.head()
```

```python
model_variables = df[['highway-mpg', 'engine-size', 'curb-weight', 'horsepower']]
model_variables.head()
```

```python
combined_df = pd.concat([drive_wheel_num, model_variables], axis=1)
```

```python
combined_df.head()
```

```python
y = df[['price']]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm
```

```python
lm.fit(combined_df, y)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
import seaborn as sns
%matplotlib inline 
```

```python
Y_hat = lm.predict(combined_df)
```

```python
width = 12
height = 10
plt.figure(figsize=(width, height))


ax1 = sns.distplot(y, hist=False, color="r", label="Actual Value")
sns.distplot(Y_hat, hist=False, color="b", label="Fitted Values" , ax=ax1)


plt.title('Actual vs Fitted Values for Price')
plt.xlabel('Price (in dollars)')
plt.ylabel('Proportion of Cars')

plt.show()
plt.close()
```

```python
print('The R-square is: ', lm.score(combined_df, y))
```

```python
from sklearn.metrics import mean_squared_error
print('The mean square error of price and predicted value using actual is: ', \
      mean_squared_error(y, Y_hat))
```

```python
import seaborn as sns
# Predicted values
y_pred = lm.predict(combined_df)

# Residuals
residuals = y - y_pred

# Plotting residuals
plt.figure(figsize=(10, 6))
sns.residplot(x=y_pred, y=residuals, lowess=True, line_kws={'color': 'red'})
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot for Multiple Linear Regression')
plt.show()
```

## Polynomial Regression Model
